```python
a=np.array([1,2,3])
b=np.array([4,5,6])
print(a+b)  # Element-wise addition
print(a-b)  # Element-wise subtraction
```

```python
matrix = np.array([[1, 2], [3, 4]])
scalar_added = matrix + 10
print(scalar_added)  # Adding a scalar to each element of the matrix
```

```python
matrix = np.array([[1, 2], [3, 4]])
print(matrix*2)
```

```python
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
matrix1@matrix2  # Element-wise multiplication
```

```python
import numpy as np 
a = np.array([10,20,30,40,50]) # Created 2 arrays
b = np.array([5,15,25,35,45]) 
#Performed basic calculations on the array for the first array
print("sum:",np.sum(a))
print("mean:",np.mean(a))
print("maximum:",np.max(a))
print("minimum:",np.min(a))
print("standard deviation:",np.std(a))
print("shape:",a.shape)
print("dimention:",a.ndim)#Explored array properties
print("data type:",a.dtype)
print("size:",a.size)
a2 = a.reshape(5,1) # Reshaped the array
print(a2)
a[2] # indexing and slicing
a[0:3]
print(a+b) # addition
print(a*b) # multiplication
```
