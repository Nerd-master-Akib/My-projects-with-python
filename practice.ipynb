```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 20, 15, 10, 5]

# Create a figure with 1 row, 2 columns
plt.subplot(1, 2, 1)  # (rows, columns, plot number)
plt.plot(x, y1, color='blue', marker='o')
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.plot(x, y2, color='red', marker='s')
plt.title("Plot 2")

plt.suptitle("Two Plots Side by Side")  # Overall figure title
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='green')
plt.title("Top Left")

plt.subplot(2, 2, 2)
plt.plot(x, y2, color='orange')
plt.title("Top Right")

plt.subplot(2, 2, 3)
plt.scatter(x, y1, color='purple')
plt.title("Bottom Left")

plt.subplot(2, 2, 4)
plt.scatter(x, y2, color='brown')
plt.title("Bottom Right")
plt.suptitle("2x2 Subplot Grid")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")

fig, axes = plt.subplots(1, 2, figsize=(12,5))  # 1 row, 2 columns

sns.boxplot(x='day', y='total_bill', data=tips, ax=axes[0])
axes[0].set_title("Box Plot")

sns.violinplot(x='day', y='total_bill', data=tips, ax=axes[1])
axes[1].set_title("Violin Plot")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

fig , axes = plt.subplots(2,2, figsize = (12,10))
sns.boxplot( x = 'day', y = 'total_bill', data = tips, ax = axes[0,0])
axes[0,0].set_title("Box Plot of Total Bill by Day")

sns.violinplot(x= 'day', y = 'total_bill', data = tips, ax = axes[0,1])
axes[0,1].set_title("Violin Plot of Total Bill by Day")

sns.countplot(x = 'day', data = tips, ax = axes[1,0])
axes[1,0].set_title("Bar Plot of Average Total Bill by Day")

sns.barplot(x = 'day', y= 'total_bill', data = tips, ax = axes[1,1])
axes[1,1].set_title("Count Plot of Days")
```
