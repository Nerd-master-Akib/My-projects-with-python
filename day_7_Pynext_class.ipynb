```python
# clean categories
def normalize_dept(s):
    s = s.astype(str).str.strip().str.lower()
    mapping = {
        'cse' : "CSE","c.s.e": "CSE",
        "eee":"EEE" , "eleceng": "EEE","e.e.e":"EEE",
        "mat":"MAT","mathmatics":"MAT",
        "sta":"STA","statistics":"STA",
        "phy":"PHY","physics":"PHY",
        "me":"MEE","mee":"MEE",
        "ipe":"IPE","industrial%production":"IPE",
        "swe":"SWE","software":"SWE"
    }
    return s.replace(mapping).str.upper()
```

```python
df["dept_clean"]=normalize_dept(df["dept"])
df.groupby("dept_clean")["student_id"].count()
```

```python
def normalize_dept(s):
    s = s.astype(str).str.strip().str.lower()
    mapping = {
        'm':"Male",'f':"Female",
        'FEMALE':"Female",'MALE':"Male",
        'male':"Male",'female':"Female"
    }
    return s.replace(mapping).str.capitalize()
```

```python
df["dept_clean"] = normalize_dept(df["gender"])
df.groupby("dept_clean")["student_id"].count()
```
