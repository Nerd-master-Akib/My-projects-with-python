```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:/Users/Win Technology/Downloads/titanic_day_06.csv")
# 1. dd date of birth column
start_date = pd.to_datetime("1910-01-01")
df["date_of_birth"] = [start_date + pd.Timedelta(days=i) for i in range(len(df))]
# 2. Filter passengers born before 1910-06-01
filtered_df = df[df["date_of_birth"] < "1910-06-01"]
# 3. Find youngest (most recent birth dates)
youngest = df.sort_values(by="date_of_birth", ascending=False).head()
# 4. Find oldest (earliest birth dates)
oldest = df.sort_values(by="date_of_birth", ascending=True).head()
print("Filtered (Born before 1910-06-01):")
print(filtered_df.head())
print("\nYoungest Passengers:")
print(youngest[["Name", "date_of_birth"]])
print("\nOldest Passengers:")
print(oldest[["Name", "date_of_birth"]])
# 5. Bar Plot: Count of Survivors vs Non-Survivors
plt.figure(figsize=(6,4))
df["Survived"].value_counts().plot(kind="bar", color=["red","green"])
plt.title("Survivors vs Non-Survivors")
plt.xlabel("Survived (0=No, 1=Yes)")
plt.ylabel("Count")
plt.show()
# 6. Pie Chart: Percentage of Males vs Females
plt.figure(figsize=(6,6))
df["Sex"].value_counts().plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.title("Percentage of Males vs Females")
plt.ylabel("")
plt.show()
# 7. Histogram: Fare Distribution
plt.figure(figsize=(6,4))
df["Fare"].plot(kind="hist", bins=30, edgecolor="black")
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
# 8. Box Plot: Fare Distribution
plt.figure(figsize=(6,4))
df["Fare"].plot(kind="box")
plt.title("Box Plot of Fare Distribution")
plt.show()
```
