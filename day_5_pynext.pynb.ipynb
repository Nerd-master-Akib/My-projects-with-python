## Merging Data

```python
import pandas as pd
df1= pd.DataFrame({"Name": ["Alice","Bob"], "Age": [25,30]})
df2= pd.DataFrame({"Name": ["Charlie","Diana"], "Age": [28,22]}) 
```

```python
df3 = pd.concat([df1, df2], ignore_index=False)
df3
```

```python
df3 = pd.DataFrame({"city":["Paris", "London"]})
pd.concat([df1, df3],axis = 1)
```

```python
import pandas as pd
import numpy as np 

# sample data with an outliner
data = pd.Series([10,12,13,14,15,100])
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

outliners = data[(data < lower_bound) | (data > upper_bound)]
outliners
```
